```python
import os
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame()

for file in filter(lambda x: x.endswith('.csv'), os.listdir('.')):
    try:
        load, api, db, result = file.split('-')
    except ValueError:
        continue
    print(file)
    _df = pd.read_csv(file)
    _df['cpu_load'] = load[:-1]
    _df['cpu_api'] = api[:-1]
    _df['cpu_db'] = db[:-1]
    df = pd.concat([df, _df])
```

```python
df['errors'] = df['summary.errors.connect'] + df['summary.errors.read'] + df['summary.errors.write'] + df['summary.errors.status'] + df['summary.errors.timeout']
df['rps'] = (df['summary.requests']) / df['summary.duration']
columns = [
    'cpu_load', 'cpu_api', 'cpu_db', 'title', 'endpoint', 'latency:percentile(90)', 
    'summary.requests', 'errors', 'rps'
]
df = df[columns]
```

```python
def plot(endpoint = '/remote'):
    t = df[(df.endpoint == endpoint)]
    return t[t['cpu_load'] == '50']

plot()
```
